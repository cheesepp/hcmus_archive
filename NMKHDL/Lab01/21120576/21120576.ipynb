# 2. Collect data from a website by parsing HTML (3p)

In this section, you are going to collect data from a website simulating the sale of Pokemon. I have prepared all the needed links in a file (`pokemon.txt`), and you have to crawl data from these links. The expected output is a `dataframe` with the following fields:

+ `SKU`: ID of Pokemon
+ `Name`: Name of Pokemon
+ `Price`: Price of Pokemon 
+ `InStock`: Quantity of Pokemons in stock
+ `Categories`: The category of Pokemon
+ `Tags`: Tags of Pokemon

Your mission is to complete function `collect_data` with `course_urls_file` as your input parameter. The output should look like `pokemon_example.csv` (I list out some examples so you can easily imagine your work).

````python
pokemon_example = pd.read_csv('pokemon_example.csv')
pokemon_example
````

````python
def collect_data(course_urls_file):
    #load paths from file
    url_file = open(course_urls_file)
    urls = url_file.readlines()
    urls_filtered = [item[:-1] for item in urls]
    
    #init empty list to store the values of each attribute.
    SKU = []
    Names = []
    Prices = []
    InStocks = []
    Categories = []
    Tags = []
    
    for url in urls_filtered:
        #TODO
        # request url to get data as string
        item = requests.get(url)
        # parse the above string into html format
        doc = BeautifulSoup(item.text, "html.parser")
        
        # get span tag for each attribute
        
        # sku tag
        spanSKU = doc.find('span', {'class': 'sku'})
        
        # price tag
        pPrice = doc.find('p', {'class': 'price'})
        span = pPrice.find('span', {'class': 'woocommerce-Price-amount amount'})
        price = span.get_text()
        price = float(price[1:])
        
        # in stock tag
        pInStock = doc.find('p', {'class': 'stock in-stock'})
        
        # categories tag
        spanCats = doc.find('span', {'class': 'posted_in'})
        aCats = spanCats.find_all('a')
        cats = []
        # add each category to categories list
        for category in aCats:
            cats.append(category.text)
        cats = ", ".join(cats)
         
        # tags tag
        spanTags = doc.find('span', {'class': 'tagged_as'})
        aTags = spanTags.find_all('a')
        tags = []
        # add each tag to tags list
        for tag in aTags:
            tags.append(tag.text)
        tags = ", ".join(tags)
        
        # append new data to attributes
        SKU.append(spanSKU.string)        
        Names.append(doc.title.string[:-11])
        Prices.append(price)
        InStocks.append(int(pInStock.text[:-9]))
        Categories.append(cats)
        Tags.append(tags)
    
    # save it to dataframe
    data = pd.DataFrame({"SKU": SKU,
                         "Name": Names,
                         "Price": Prices, 
                         "InStock": InStocks,
                         "Categories": Categories,
                         "Tags": Tags})
    
    return data
````

````python
#TEST
data_pokemon = collect_data("pokemon.txt") 
assert data_pokemon.shape == (755, 6)
````

````python
#Save to csv file with name pokemon.csv
data_pokemon.to_csv('pokemon.csv')
````

````python
BASE_URL = 'http://api.worldbank.org/v2/'
COUNTRIES = ["US", "IN", "CN", "JP", "CA", "GB", "ZA"]
INDICATORS = ['SP.POP.TOTL', 
             'SP.POP.TOTL.FE.IN', 
             'SP.POP.TOTL.MA.IN',
             'SP.DYN.CBRT.IN', 
             'SP.DYN.CDRT.IN',
             'SP.DYN.LE00.MA.IN',
             'SP.DYN.LE00.FE.IN',
             'SE.PRM.ENRR',
             'SE.TER.ENRR',
             'SE.PRM.CMPT.ZS',
             'SE.ADT.1524.LT.ZS']

#TODO (option)
# If you need other initializations

# group api attributes of each country to a list
apiUS = [BASE_URL + 'countries' + '/us' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiIN = [BASE_URL + 'countries' + '/in' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiCN = [BASE_URL + 'countries' + '/cn' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiJP = [BASE_URL + 'countries' + '/jp' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiCA = [BASE_URL + 'countries' + '/ca' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiGB = [BASE_URL + 'countries' + '/gb' + '/indicators/' + endpoint for endpoint in INDICATORS]
apiZA = [BASE_URL + 'countries' + '/za' + '/indicators/' + endpoint for endpoint in INDICATORS]

# create a dictionary for easily processing
countries_api = ({'us': apiUS, 
                  'in': apiIN,
                  'cn': apiCN, 
                  'jp': apiJP, 
                  'ca': apiCA,
                  'gb': apiGB,
                  'za': apiZA})

# change each country to lower case for easily parsing url
COUNTRIES = [country.lower() for country in COUNTRIES]
````

````python
# create attribute list
attributes = ['Total Population', 
        'Female Population',
        'Male Population', 
        'Birth Rate',
        'Death Rate', 
        'Male life expectancy', 
        'Female life expectancy', 
        'School enrollment, primary', 
        'School enrollment, tertiary', 
        'Primary completion rate', 
        'Literacy rate', 
        'Year',
        'Country']

def collect_data(countryCode, per_page, start_year, end_year):
    # create empty data frame
    collect = pd.DataFrame()
    
    # check if year and country were filled
    is_fill = False
    years = []
    countries = []
    for api_index in range(0, len(countries_api[countryCode])):
        api = countries_api[countryCode][api_index]
        # request api to get data
        url = f'{api}?per_page={per_page}&format=json'
        response = requests.get(url)
        if response.status_code == 200:
            json_data = response.json()
            # store attribute to list
            attribute = []
            # iterate each data in json body
            for data in json_data[1]:
                # get year
                year = int(data["date"])
                # get country
                country = data["countryiso3code"]
                if year >= start_year and year <= end_year:
                    value = str(data["value"])
                    if value == 'None':
                        attribute.append('None')
                    else:
                        attribute.append(float(value))
                    if is_fill == False:
                        years.append(year)
                        countries.append(country)

                else:
                    pass
                
        is_fill = True
        # store attribute to the collect data frame
        collect[attributes[api_index]] = attribute
    # store year and country to data frame
    collect['Year'] = years
    collect['Country'] = countries
    
    return collect        
````

````python
def Generate_Countries_Dataset(countryCode_list):
    data = pd.DataFrame()
    for countryCode in countryCode_list:
        data = pd.concat([data, collect_data(countryCode = countryCode, per_page = 100, start_year = 2000, end_year = 2022)], axis=0)
    return data
````

````python
#TEST
data_countries = Generate_Countries_Dataset(COUNTRIES)
assert data_countries.shape == (161, 13)
````

````python
# Save to csv file with name coutries.csv
data_countries.to_csv('countries.csv')
````
