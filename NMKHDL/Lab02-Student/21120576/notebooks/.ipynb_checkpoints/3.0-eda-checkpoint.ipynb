**Question 01**: How does the correlation between features change in the top 10 countries with the LARGEST population?

When we have an answer for this question, we might see how features interaction with each other when the conutry population is large.

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION)

# get top population of each country in 2022
groupby_df = cleaned_df.query('Year == 2022').groupby(["Country"])["Total Population"].nlargest(1)

#get temporary cleaned data for manipulate
temp_df = cleaned_df.copy()

# merge the groupby data into the temp data 
df_top10_largest = temp_df.merge(groupby_df.reset_index(), on=['Country', 'Total Population'])

# drop the redundant feature
df_top10_largest = df_top10_largest.drop('level_1', axis='columns')
# sort population in descending order
df_top10_largest.sort_values("Total Population", ascending=False).head(10)
```

```python
# YOUR CODE HERE
df_copy = df_top10_largest.sort_values("Total Population", ascending=False).drop(['Year', 'Country'], axis='columns').head(10)
corr_matrix = df_copy.corr()
sns.heatmap(corr_matrix)
plt.show()
```

**YOUR ANSWER**
- As show on the heatmap, the correlation between features is weak, just have some correlation coefficient.
- The major is no-correlated because data in 2022 were filled by us, not the real value so in some column, all elements have the same value.

**Question 02**: How does the correlation between features change in the top 10 countries with the SMALLEST population?

When we have an answer for this question, we might see how features interaction with each other when the conutry population is small.

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION) 

groupby_df = cleaned_df.query('Year == 2022').groupby(["Country"])["Total Population"].nsmallest(1)
temp_df = cleaned_df.copy()

df_top10_smallest = temp_df.merge(groupby_df.reset_index(), on=['Country', 'Total Population'])
df_top10_smallest = df_top10_smallest.drop('level_1', axis='columns')
df_top10_smallest.sort_values("Total Population", ascending=True).head(10)
```

```python
# YOUR CODE HERE
df_copy = df_top10_smallest.sort_values("Total Population", ascending=True).drop(['Year', 'Country'], axis='columns').head(10)
corr_matrix = df_copy.corr()

sns.heatmap(corr_matrix, annot=False)
plt.show()
```

**YOUR ANSWER**
- Like above top 10 largest population countries, this one is not so different.

**How we answer this question**: we will consider the Average Birth Rate and Death Rate for countries across all the year.

### Pre-processing

### Analyzing to get the answer

Step 01: construct Average Birth Rate dataframe for each country.

```python
# Create a new dataframe
avg_birthrate_df = pd.DataFrame()

# YOUR CODE HERE
# calculate average birth rate
avg_birthrate_df['Average Birth Rate'] = cleaned_df.groupby('Country')['Birth Rate'].mean()
# reset index to become a feature column
avg_birthrate_df = avg_birthrate_df.reset_index()
avg_birthrate_df
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE
# sort values to get the top average birth rate
avg_birthrate_df = avg_birthrate_df.sort_values('Average Birth Rate', ascending=False).head(10)
sns.barplot(avg_birthrate_df, x="Average Birth Rate", y="Country")
```

Step 02: construct Average Death Rate dataframe for each country.

```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# YOUR CODE HERE
# calculate average birth rate
avg_deathrate_df['Average Death Rate'] = cleaned_df.groupby('Country')['Death Rate'].mean()
# reset index to become feature column
avg_deathrate_df = avg_deathrate_df.reset_index()
avg_deathrate_df
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE

# sort values to get the top average death rate
avg_deathrate_df = avg_deathrate_df.sort_values('Average Death Rate', ascending=False).head(10)
sns.barplot(avg_deathrate_df, x="Average Death Rate", y="Country")
```

## Making your own question 01? (1.5 points)

**Question:** What is the trend in average life expectancy from 1995 to 2022 in Asia?

**YOUR ANSWER HERE**

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
# create empty dataframe
highest_life_expec_df = pd.DataFrame()

# calculate average male and female life expectancy
highest_life_expec_df['Average male life expectancy'] = cleaned_df.groupby('Year')['Male life expectancy'].mean() 
highest_life_expec_df['Average female life expectancy'] = cleaned_df.groupby('Year')['Female life expectancy '].mean() 
highest_life_expec_df
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
# create line plot for trending graph
sns.lineplot(data=highest_life_expec_df, x="Year", y="Average male life expectancy", label='Average male life expectancy')
sns.lineplot(data=highest_life_expec_df, x="Year", y="Average female life expectancy", label='Average female life expectancy')
plt.title("Average life expectancy by gender from 1995 to 2022 in asia")
plt.show()
```

**ANSWER:** As we can see in the trending graph, in both genders have the same direction:
- The average life expectation of asian tends to increase strongly from 1995 to roughly 2017-2018. After that, both go down very fast
- The average female life expectancy has a huge gap compare to male cross all the year, the difference is approximately 6.

**The average life expectancy can indicate economic and social development, healthcare quality, and living conditions for the population in the region, but it should also be assessed in the context of specific factors for each country or region to gain a deeper understanding of their specific life expectancy situations.**

## Making your own question 02? (1.5 points)

**Question:** Which countries in ASEAN has the highest and second-highest GDP? *(in the given dataset)*

**YOUR ANSWER HERE**

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 

# create asean countries list
ASEAN = ['Brunei Darussalam', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Philippines', 'Singapore', 'Thailand', 'Vietnam', 'Myanmar']

# check clean data which has country is in the asean list, add it to asean dataframe
asean_df = cleaned_df.loc[cleaned_df['Country'].isin(ASEAN)]

# calculate GDP for each country cross all the years (1995 - 2022)
asean_avg_gdp = asean_df.groupby('Country')['GDP in USD'].mean()

# reset country index to be a feature column
asean_avg_gdp = asean_avg_gdp.reset_index()

# sort GDP in descending order
asean_avg_gdp = asean_avg_gdp.sort_values('GDP in USD', ascending=False)
# convert to billion
asean_avg_gdp['GDP in USD'] = asean_avg_gdp['GDP in USD'] / 10**9
# rename GDP to understand easily
asean_avg_gdp.rename(columns={'GDP in USD': 'Average GDP in USD (billion)'}, inplace=True)
asean_avg_gdp
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
sns.barplot(asean_avg_gdp, x="Average GDP in USD (billion)", y="Country")
```

**ANSWER:** 
- As show in the bar plot, Indonesia has the highest GPD in ASEAN which is over 600 billion USD, far ahead of other countries.
- Right after that it is Thailand which is over 300 billion, less than a half compare to Indonesia.
