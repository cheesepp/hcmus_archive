```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Đọc Dataframe từ file csv, với các cột như sau
field_names = ['Region', 'Country', 'HIV Treatment Rate', 
               'GNI per capita (USD)', 'GDP growth (annual %)',
               'Adult literacy rate','GDP per capita growth']
world_df = pd.read_csv('world_data.csv',skiprows=1,names=field_names)
world_df
```

```python
# Vẽ barplot cho các Region
# TODO
# Đếm số đất nước theo mỗi khu vực
grouped_regions = world_df.groupby('Region').count().sort_values('Country', ascending=False)
# Vẽ đồ thị bar chart
plt.barh(grouped_regions.index, grouped_regions['Country'])
```

```python
# Vẽ boxplot cho 'GNI per capita (USD)'
# TODO
# Loại bỏ những dòng na
data = world_df['GNI per capita (USD)'].dropna()
# Vẽ boxplot
plt.boxplot(data)
```

```python
# Vẽ violinplot cho 'GNI per capita (USD)'
data = world_df['GNI per capita (USD)'].dropna()
plt.violinplot(data)
# TODO
```

```python
# Vẽ boxplot của 'GDP growth (annual %)' thuộc các Region lên cùng một biểu đồ
# Lấy khu vực
unique_regions = world_df['Region'].dropna().unique()

# Lấy gdp từng khu vực
gdp_growth = [world_df[world_df['Region'] == r]['GDP growth (annual %)'].dropna().tolist() for r in unique_regions]

# Vẽ boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(gdp_growth, labels=unique_regions)
plt.title('Boxplot of Values by Class')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.ylabel('GDP growth (annual %)')
plt.show()
# TODO
```

```python
# Vẽ histogram của 'GNI per capita (USD)'
# TODO
plt.hist(world_df['GNI per capita (USD)'], bins=20, edgecolor='white')
```

Gọi các vùng sau là bán cầu bắc (Northern)
- 'Europe & Central Asia'
- 'North America'
- 'East Asia & Pacific'
  
Gọi các vùng sau là bán cầu nam (Southern)
- 'Sub-Saharan Africa'
- 'Middle East & North Africa'
- 'South Asia'
- 'Latin America & Caribbean'

```python
# Hãy vẽ 2 histogram thể hiện sự phân bố của cột 'GNI per capita (USD)'
# cho bán cầu bắc và bán cầu nam lên cùng một bản đồ
# TODO
# Lấy những khu vực phía bắc
northern = ['Europe & Central Asia', 'North America', 'East Asia & Pacific']
# Lấy những khu vực phía nam
southern = ['Sub-Saharan Africa', 'Middle East & North Africa', 'South Asia', 'Latin America & Caribbean']
# Lấy những giá trị GNI từng khu vực ở trên
northern_df = world_df[world_df['Region'].isin(northern)]['GNI per capita (USD)'].tolist()
southern_df = world_df[world_df['Region'].isin(southern)]['GNI per capita (USD)'].tolist()

# Vẽ đồ thị histogram
kwargs = dict(histtype='barstacked',bins=20,edgecolor='black', density=True, alpha = 0.3)

plt.hist(northern_df, label='Northern', **kwargs)
plt.hist(southern_df, label='Southern', **kwargs)
plt.legend()
```

```python
# Hãy vẽ scatter plot thể hiện mối quan hệ giữa
#'GNI per capita (USD)' và 'Adult literacy rate'
# TODO
plt.scatter(world_df['GNI per capita (USD)'], world_df['Adult literacy rate'])
```

```python
# Hãy vẽ scatter plot thể hiện mối quan hệ giữa
#'GNI per capita (USD)' và 'Adult literacy rate'
# cho mỗi Region với màu khác nhau
# TODO
# Lấy các khu vực
unique_regions = world_df['Region'].dropna().unique()
# Vẽ từng scatter plot tương ứng với từng khu vực
for region in unique_regions:
    plt.scatter(world_df[world_df['Region'] == region]['GNI per capita (USD)'], world_df[world_df['Region'] == region]['Adult literacy rate'], label=region)
plt.legend()
```

```python
# Hãy vẽ scatter plot thể hiện mối quan hệ giữa
#'GNI per capita (USD)' và 'Adult literacy rate'
# Cùng với hai histogram ở hai trục
# TODO
# Tạo ax với gridspec
fig = plt.figure(figsize=(8, 6))

grid = plt.GridSpec(6, 6, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:, :5])
y_hist = fig.add_subplot(grid[1:, 5], sharey=main_ax, yticklabels=[])
x_hist = fig.add_subplot(grid[0, :5], sharex=main_ax, xticklabels=[])
# vẽ scatter points lên trục chính 
main_ax.scatter(world_df['GNI per capita (USD)'], world_df['Adult literacy rate'], alpha=0.3)
main_ax.set_xlabel('GNI per capita (USD)')
main_ax.set_ylabel('Adult literacy rate')

# vẽ histogram gắn trên mỗi trục
x_hist.hist(world_df['GNI per capita (USD)'], 20, histtype='barstacked',
            orientation='vertical', color='blue', edgecolor='black')
y_hist.hist(world_df['Adult literacy rate'], 20, histtype='barstacked',
            orientation='horizontal', color='red', edgecolor='black')
```
