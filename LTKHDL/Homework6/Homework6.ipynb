```python
import pandas as pd
import numpy as np
```

```python
# Đọc Dataframe từ file csv, với các cột như sau
field_names = ['State', 'Sex', 'Year', 'Name', 'Count']
babynames_df = pd.read_csv('STATE.CA.TXT',header=None,names=field_names)
```

```python
# Show 10 dòng ngẫu nhiên
babynames_df.sample(10)
```

```python
# Hiển thị DF chỉ gồm female 
babynames_df[babynames_df['Sex'] == 'F']
```

```python
# Hiển thị DF gồm những người male, sinh trước 2000
babynames_df[(babynames_df['Sex'] == 'M') & (babynames_df['Year'] < 2000)]
```

```python
# Hiển thị 3 babynames với count > 250
babynames_df[(babynames_df['Count'] > 250)].sample(3)
```

```python
# Hiển thị các babynames có tên là Bella, Alex, Narges, hoặc Lisa
lst = ['Bella', 'Alex', 'Narges', 'Lisa']
babynames_df[babynames_df['Name'].isin(lst)]
```

```python
# Hiển thị với những Name bắt đầu bằng chữ N
babynames_df[babynames_df['Name'].str.startswith('N')]
```

```python
# Thêm cột name_lengths với giá trị là số lượng ký tự của cột Name
babynames_df['name_lengths'] = babynames_df['Name'].str.len()
babynames_df
```

```python
# Hãy cho biết số lượng baby có tên 'Alice' trung bình hàng năm là bao nhiêu
average_Alice = babynames_df[babynames_df['Name'] == 'Alice']['Count'].mean()
average_Alice
```

```python
# Hiển thị tất cả các tên trong DF, mỗi tên chỉ hiển thị 1 lần
list(babynames_df['Name'].unique())
```

```python
# Tìm babynames phổ biến nhất trong năm 2020
year_2020 = babynames_df[babynames_df['Year'] == 2020]
# Đếm số lượng xuất hiện mỗi tên và chọn ra những tên xuất hiện nhiều nhất
year_2020['Name'].value_counts().nlargest()
```
