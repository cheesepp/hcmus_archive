## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
import re

contain_num = ["Call me at 382-384-3840.",  "my number is (510) 849-3519. Call me!", "my number is 510-849-35192", "here’s my number: 510-849.3519"]

# create pattern for number
pattern = r'^(\(\d{3}\)\s?\d{3}-\d{4}|\d{3}-\d{3}-(\d{4})*)$'

# check each string in 
for i in range(len(contain_num)):
    if re.search(pattern, contain_num[i]) != None:
        print(f'\"{contain_num[i]}\"'+ " contains phone number!")
    else:
        print(f'\"{contain_num[i]}\"'+ " not contains phone number!")
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
text = "03/12/2018, 03.13.18, 03/14/2018, 03:15:2018"

# define a pattern to match various date formats
pattern = r'(\d{2})[/.:](\d{2})[/.:](\d{2,4})'

# define a function to replace matched patterns with dashes
def replace_date(match):
    return f"{match.group(1)}-{match.group(2)}-{match.group(3)}"

# Use re.sub to perform the replacement
formatted_text = re.sub(pattern, replace_date, text)

print(formatted_text)
````

## Exercise 3

Consider the first five sentences of the novel “Little Women” below. Extract the spoken dialog from each sentence.

````python
text = '''
"Christmas won't be Christmas without any presents," grumbled Jo, lying on the rug.
"It's so dreadful to be poor!" sighed Meg, looking down at her old dress.
"I don't think it's fair for some girls to have plenty of pretty things, and other girls nothing at all," added little Amy, with an injured sniff.
"We've got Father and Mother, and each other," said Beth contentedly from her corner.
The four young faces on which the firelight shone brightened at the cheerful words, but darkened again as Jo said sadly, "We haven't got Father, and shall not have him for a long time."
'''
print(text)
````

````python
# split the string into the elements of a list
text = text.strip().split('\n')
````

````python
# create pattern to find the dialog
pattern = r'"(.*?)"'
for t in text:
    print(re.findall(pattern, t))
````

## Exercise 4

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this
```
Name:
"MR. JAMES NGOLA." 
"Mr. Ben Suleman" 
"PRINCE OBONG ELEME" 
"PRINCE OBONG ELEME" 
"Maryam Abacha" 
```
```
Email address:
"james_ngola2002@maktoob.com"
"bensul2004nng@spinfinder.com"
"obong_715@epatra.com"
"obong_715@epatra.com"
"m_abacha03@www.com"
```
2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004 


````python
import re
import datetime as dt
fh = open(r"email_test.txt", "r").read()

# YOUR CODE HERE
# 1. Extract name and email address of sender
# get name and email from the "From: ...." line
name_email_pattern = r'From: ("[A-Za-z\s.]+") <([A-Za-z0-9_@.]+)>'
res = re.findall(name_email_pattern, fh)

print('Name:')
for name, email in res:
    print(name)
    
print('Email:')
for name, email in res:
    print(f'"{email}"')

# 2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004
# get subject
sub_pattern = r'Subject: (.*)'
# get date
date_pattern = r'Date: [A-Za-z]+, (\d+ [A-Za-z]+ \d{4})'
# find all subjects
subjects = re.findall(sub_pattern, fh)

# save subjects which is sent from 01/06/2003 to 13/05/2004 
subjects_in_range = []
# parse start date and end date
s = dt.datetime.strptime("01/06/2003", "%d/%m/%Y") 
e = dt.datetime.strptime("13/05/2004", "%d/%m/%Y") 

for index, date in enumerate(re.finditer(date_pattern, fh)):
    # Parse the date string into a datetime object
    parse_date = dt.datetime.strptime(date.group(1), "%d %b %Y")
    if parse_date >= s and parse_date <= e:
        # add in-range subject to list
        subjects_in_range.append(subjects[index])
# print out
subjects_in_range
````
